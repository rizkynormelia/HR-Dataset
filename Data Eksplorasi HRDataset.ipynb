```python
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
import itertools
```

```python
# visualizations of categorical variable
categorical=['MaritalDesc','Sex','CitizenDesc','RaceDesc', 'Department','Performance Score']
fig=plt.subplots(figsize=(10,20))
length=len(categorical)
for i,j in itertools.zip_longest(categorical,range(length)):
    plt.subplot(np.ceil(length/2),3,j+1)
    plt.subplots_adjust(hspace=.6, wspace=.5)
    sns.countplot(x=i,data = hrdata)
    plt.xticks(rotation=90)
    plt.title("Total Employee")
```

1. Most of the employees marital status are single and married
2. There's no wide gap between male and female employee
3. Most of the employees have US citizen and dominated by white race
4. The highest employees are in production department followed by IT/IS and sales department
5. There are more employees who have good work performance

```python
# visualizations of department and performance score ID by sex
categorical=['Department','Perf_ScoreID'] # here I have removed left to see who is leaving cpmpany
fig=plt.subplots(figsize=(10,15))# to define the size of figure
length=len(categorical) # no of categorical and ordinal variable
for i,j in itertools.zip_longest(categorical,range(length)): # itertools.zip_longest for to execute the longest loop
    plt.subplot(np.ceil(length),2,j+1) # this is to plot the subplots like as 2,2,1 it means 2x2 matrix and graph at 1 
    plt.subplots_adjust(hspace=.5) # to adjust the distance between subplots
    sns.countplot(x=i,data=hrdata,hue="Sex") # To plot the countplot of variable with hue left
    plt.xticks(rotation=90) # to rotate the xticks by 90 such that no xtixks overlap each other
    plt.title("Total employee") # to plot the title of graph
```

1. Production department is dominated by female employees
2. There's no male employee in executive office department
2. The good performance score (3,4,5) mostly filled by female
